```python
import numpy as np
import matplotlib.pyplot as plt
# Scikit-Fuzzy es una colección de algoritmos de lógica difusa escritos en Python
import skfuzzy as fuzzb
```

2. Mediante la siguiente línea, creamos el universo del discurso. ¿Qué hace en
concreto el método “linspace” del objeto “np” (numpy)? Búscalo en internet.

```python
X = np.linspace(start=0, stop=75, num=75, endpoint=True, retstep=False)
```

3. Podemos hacer operaciones con conjuntos creando previamente dos funciones
de membresía de tipo triangular mediante los métodos que vienen con “scikit”.

```python
# Creamos valores para dos conjuntos correspondientes a edades
edades_1 = [0, 25, 50]
edades_2 = [25, 50, 75]

# Creamos sendas funciones de membresía con forma triangular
jovenes = fuzzb.membership.trimf(X, edades_1)
mediana_edad = fuzzb.membership.trimf(X, edades_2)
```

4. Ahora, representamos ambas funciones:

```python
plt.figure()  # prepara el plano de representación
# Función triangular “Jóvenes”
plt.subplot(3, 2, 1)  # ubica la 1ª gráfica en el plano
plt.plot(X, jovenes)  # dibuja la 1ª función de membresía
plt.title("Jóvenes")
plt.grid(True)
# Función triangular “Mediana Edad”
plt.subplot(3, 2, 2)  # ubica la 2ª gráfica en el plano
plt.plot(X, mediana_edad)  # dibuja la 2ª función de membresía
plt.title("Mediana Edad")
plt.grid(True)
plt.show()
```

En las siguientes líneas vamos a representar diferentes operaciones que hemos
revisado a través de las diapositivas. Como ejemplo, a continuación tienes la función
de unión.

5. Operación de unión:

```python
# 1. Unión = max(µA(x), µB(x))
union = fuzzb.fuzzy_or(X, jovenes, X, mediana_edad)[1]
plt.subplot(2, 2, 1)
plt.plot(X, union)
plt.title("Unión")
plt.grid(True)
plt.show()
```

6. Operación de intersección:

```python
# 2. Intersección = min(µA(x), µB(x))
intersection = fuzzb.fuzzy_and(X, jovenes, X, mediana_edad)[1]
plt.subplot(2, 2, 1)
plt.plot(X, intersection)
plt.title("Intersección")
plt.grid(True)
plt.show()
```

7. Operación de complemento (de A):

```python
# 3. Complemento (A) = (1- min(µA(x))
complement = fuzzb.fuzzy_not(jovenes)
plt.subplot(3, 2, 2)
plt.plot(X, complement)
plt.title("Complemento")
plt.grid(True)
plt.show()
```

8. Operación de diferencia (A/B):

```python
# 4. Diferencia (A/B) = min(µA(x),(1 - µB(x)))
complement_of_b = fuzzb.fuzzy_not(mediana_edad)

division = fuzzb.fuzzy_and(X, jovenes, X, complement_of_b)[1]
plt.subplot(2, 2, 1)
plt.plot(X, division)
plt.title("Diference (A / B)")
plt.grid(True)
plt.show()
```

9. Operación de producto algebraico:

```python
# 5. Producto algebraico = (µA(x) * µB(x))
product = jovenes * mediana_edad
plt.subplot(2, 2, 1)
plt.plot(X, product)
plt.title("Product (A * B)")
plt.grid(True)
plt.show()
```
