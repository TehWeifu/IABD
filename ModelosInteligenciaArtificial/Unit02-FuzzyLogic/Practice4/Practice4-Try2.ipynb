```python
import numpy as np
from fuzzy import *
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
# Range of food quality
food_quality_universe_discourse = np.arange(0, 11, 1)

# Range of service quality
service_quality_universe_discourse = np.arange(0, 11, 1)

# Range of tip percentage (from 5% to 25%)
tip_percent_universe_discourse = np.arange(5, 26, 1)
```

```python
# Membership function for food quality
food_quality_membership_low = trimf(food_quality_universe_discourse, [-1, 0, 5])
food_quality_membership_normal = trimf(food_quality_universe_discourse, [0, 5, 10])
food_quality_membership_high = trimf(food_quality_universe_discourse, [5, 10, 10])

# Graphs for the quality of the food
plt.plot(food_quality_universe_discourse, food_quality_membership_low, label="Low")
plt.plot(food_quality_universe_discourse, food_quality_membership_normal, label="Normal")
plt.plot(food_quality_universe_discourse, food_quality_membership_high, label="High")
plt.legend(loc='best')
plt.xlabel('x= Food quality')
plt.ylabel('$\mu (x)$')
plt.show()
```

```python
# Membership function for service quality
service_quality_membership_low = trimf(service_quality_universe_discourse, [-1, 0, 5])
service_quality_membership_normal = trimf(service_quality_universe_discourse, [0, 5, 10])
service_quality_membership_high = trimf(service_quality_universe_discourse, [5, 10, 10])

# Graphs for the quality of the service
plt.plot(service_quality_universe_discourse, service_quality_membership_low, label="Low")
plt.plot(service_quality_universe_discourse, service_quality_membership_normal, label="Normal")
plt.plot(service_quality_universe_discourse, service_quality_membership_high, label="High")
plt.legend(loc='best')
plt.xlabel('x= Service quality')
plt.ylabel('$\mu (x)$')
plt.show()
```

```python
# Membership function for service quality
tip_percent_membership_low = trimf(tip_percent_universe_discourse, [4, 5, 13])
tip_percent_membership_normal = trimf(tip_percent_universe_discourse, [5, 13, 25])
tip_percent_membership_high = trimf(tip_percent_universe_discourse, [13, 25, 25])

# Graphs for the quality of the service
plt.plot(tip_percent_universe_discourse, tip_percent_membership_low, label="Low")
plt.plot(tip_percent_universe_discourse, tip_percent_membership_normal, label="Normal")
plt.plot(tip_percent_universe_discourse, tip_percent_membership_high, label="High")
plt.legend(loc='best')
plt.xlabel('x= Tip percent')
plt.ylabel('$\mu (x)$')
plt.show()
```

```python
# Introduce values for the input variables
food_quality = 8
service_quality = 7

# Calculamos el grado de membresía de cierto valor de entrada para la variable 'calidad de comida' 
cal_nivel_b = fuzz.interp_membership(food_quality_universe_discourse, food_quality_membership_low, food_quality)
cal_nivel_m = fuzz.interp_membership(food_quality_universe_discourse, food_quality_membership_normal, food_quality)
cal_nivel_a = fuzz.interp_membership(food_quality, food_quality_membership_high, food_quality)
```
