```python
def has_banner_class_and_hello_world(tag):
    class_condition = tag.get('class') is not None and "banner" in tag.get('class')
    string_condition = tag.string == "Hola mundo"

    return class_condition and string_condition


soup.find_all(has_banner_class_and_hello_world)
```
