```python
import requests
from bs4 import BeautifulSoup
import spacy
import html2text

# Paso 1: Scraping de la página de historia
url = "https://books.toscrape.com/catalogue/category/books/history_32/index.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extraer información relevante, por ejemplo, títulos y descripciones de libros
libros = []
for libro_tag in soup.find_all('h3'):
    titulo = libro_tag.a['title']
    link = libro_tag.a['href'].replace("../../..", "")
    libro_url = f"https://books.toscrape.com/catalogue{link}"
    libro_response = requests.get(libro_url)
    libro_soup = BeautifulSoup(libro_response.text, 'html.parser')
    descripcion_element = libro_soup.find("article", class_="product_page")
    #descripcion = html2text.html2text(str(descripcion_element.find_all('p')[3]))
    descripcion = descripcion_element.find_all('p')[3].get_text()

    libros.append({
        'titulo': titulo,
        'descripción': descripcion
    })

print(libros)
```

```python
import collections

# Paso 2: Etiquetado morfológico con Spacy
nlp = spacy.load("en_core_web_sm")

# Procesar las descripciones de los libros y extraer sustantivos

total_sustantivos = list()
for libro in libros:
    descripcion = libro['descripción']
    doc = nlp(descripcion)

    # Extraer sustantivos y acumularlos en una lista
    sustantivos = [token.text for token in doc if token.pos_ == "NOUN"]
    total_sustantivos += sustantivos

cuentaSustantivos = collections.Counter(total_sustantivos)
cuentaSustantivos.most_common()[:20]
```

```python
import collections

# Paso 2: Etiquetado morfológico con Spacy
nlp = spacy.load("en_core_web_sm")

# Procesar las descripciones de los libros y extraer 2-gramas de tipo adjetivo + sustantivo

total_2gramas = list()
for libro in libros:
    descripcion = libro['descripción']
    doc = nlp(descripcion)

    # Extraer 2-gramas de tipo nombre + adjetivo y acumularlos en una lista
    for i in range(len(doc) - 1):
        if doc[i].pos_ == "ADJ" and doc[i + 1].pos_ == "NOUN":
           dos_grama = f"{doc[i].text} {doc[i + 1].text}"
           total_2gramas.append(dos_grama)


cuenta_2gramas = collections.Counter(total_2gramas)
cuenta_2gramas.most_common()[:20]
```
