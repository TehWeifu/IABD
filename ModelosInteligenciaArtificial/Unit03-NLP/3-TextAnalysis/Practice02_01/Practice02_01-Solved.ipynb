## **PLN - PRÁCTICA 1.2: Procesamiento de información médica mediante expresiones regulares**

### Parte 1: Extracción de Datos con Expresiones Regulares

```python
import re
import pandas as pd


dir_fich = "/content/"
lista_fich = list()
lista_informes = list()

for i in range(1, 21):
    nombre_archivo = f"informe{i}.txt"
    lista_fich.append(nombre_archivo)


for file in lista_fich:

  with open(dir_fich + file, 'r') as file:
    informe_medico = file.read()

  nombre_paciente = re.search(r'Paciente: (.+)', informe_medico).group(1)
  fecha_informe = re.search(r'Fecha: (\d{4}-\d{2}-\d{2})', informe_medico).group(1)
  edad = int(re.search(r'Edad: (\d+)', informe_medico).group(1))
  sexo = re.search(r'Sexo: (.+)', informe_medico).group(1)
  historial_codigo = re.search(r'Código: (.+)', informe_medico).group(1)
  sintomas = re.search(r'Síntomas: (.+)', informe_medico).group(1)
  enfermedad = re.search(r'Patología: (.+)', informe_medico).group(1)
  gravedad = re.search(r'Gravedad: (.+)', informe_medico).group(1)
  tratamiento = re.search(r'Tratamiento: (.+)', informe_medico).group(1)
  proxima_cita = re.search(r'Próxima Cita: (\d{4}-\d{2}-\d{2})', informe_medico).group(1)

  informe_dict = {
      "nombre_paciente": nombre_paciente,
      "fecha_informe": fecha_informe,
      "edad": edad,
      "sexo": sexo,
      "historial_codigo": historial_codigo,
      "sintomas": sintomas,
      "enfermedad": enfermedad,
      "gravedad": gravedad,
      "tratamiento": tratamiento,
      "proxima_cita": proxima_cita
  }

  lista_informes.append(informe_dict)
```
