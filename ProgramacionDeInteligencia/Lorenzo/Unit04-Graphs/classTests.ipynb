```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from matplotlib.ticker import MaxNLocator
```

```python
figure = plt.figure(figsize=(3, 6))
plt.suptitle("MY GRAPHS")

axes1 = figure.add_subplot(4, 2, 1)
axes1.set_title("Fruits")

axes2 = figure.add_subplot(4, 2, 3)
axes2.set_title("Vegetables")

axes3 = figure.add_subplot(2, 2, 2)
axes3.set_title("Pipas")

axes4 = figure.add_subplot(2, 1, 2)
axes4.set_title("Cool numbers")

figure.tight_layout()
```

```python
iris = load_iris()

figure = plt.figure()
figure.set_facecolor("#CCCCCC")
axes = figure.add_subplot()

axes.set_facecolor("#F0F7FF")
axes.grid(visible=True, which='major', axis='both', color="#FFFFFF", linewidth=1)
axes.set_axisbelow(True)

scatter = axes.scatter(iris.data[:, 2], iris.data[:, 0], c=iris.target)

axes.set_xlabel(iris.feature_names[2], fontsize=15, labelpad=20, color="#003B80")
axes.set_ylabel(iris.feature_names[0], fontsize=15, labelpad=20, color="#003B80")
axes.set_title("Flowers", fontsize=20, pad=30, color="#003B80")

# Set x and y limits
axes.set_xlim([0, 8])
axes.set_ylim([0, 10])

# Set x and y ticks
# axes.set_xticks(np.arange(0, 11, 1))  #create ticks from minimum to maximum with step 1
# axes.set_yticks(np.arange(0, 11, 1))  #create ticks from minimum to maximum with step 1
axes.xaxis.set_major_locator(MaxNLocator(integer=True))
axes.yaxis.set_major_locator(MaxNLocator(integer=True))

# Here we create legend
labels = np.unique(iris.target)
legends = axes.legend(handles=scatter.legend_elements()[0], labels=iris.target_names.tolist())

plt.show()
```
