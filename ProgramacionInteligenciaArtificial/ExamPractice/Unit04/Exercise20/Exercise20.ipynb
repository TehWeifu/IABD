```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("fifa.csv")

figure = plt.figure(figsize=(10, 6))
axe = figure.add_subplot()

sns.scatterplot(x='work_rate',
                y=df['wage_eur'],
                hue="league_rank",
                data=df,
                palette=["green", "red", "coral", "blue"],
                legend="full",
                alpha=0.4,
                ax=axe
                )

max_wage_eur = df.groupby("work_rate")["wage_eur"].max()
#Making a line plot of max wages
sns.lineplot(data=max_wage_eur,
             ax=axe,
             color="grey")
axe.tick_params(axis="x", rotation=90)
axe.set_xlabel("Work Rate")
axe.set_ylabel("Wage EUR")
figure.suptitle("Relationship between work rate and wage by league rank", fontsize=18)
```
