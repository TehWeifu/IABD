- axes:Axes donde se dibujará
- x:Array con los valores de la X
- y:Array con los valores de la Y
- target:Array con el Tipo de flor
- label_x:Label del eje X
- label_y:Label del eje Y
- 
Esa función deberá hacer la misma gráfica del ejercicio anterior.

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
```

```python
def plot_single_scatter(axes, x, y, target, label_x, label_y):
    scatter = axes.scatter(x, y, c=target, cmap='viridis')

    axe.set_xlabel(label_x)
    axe.set_ylabel(label_y)
    axe.legend(handles=scatter.legend_elements()[0], labels=np.unique(target).tolist())
```

```python
iris = load_iris()
sepal_length = iris.data[:, 0]
petal_length = iris.data[:, 2]
```

```python
figure = plt.figure(figsize=(8, 8))
axe = figure.add_subplot()

plot_single_scatter(axe, sepal_length, petal_length, iris.target, 'Largo Sepalo', 'Largo Petalo')
```
