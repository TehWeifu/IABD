Muestra un gráfico KDE con la distribución de cada tipo de flor en función del largo del sépalo

La etiqueta del eje X sea "largo sépalo"

```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
feature_chosen = 0

iris = load_iris()
sepal_length = iris.data[:, feature_chosen]
```

```python
figure = plt.figure(figsize=(6, 6))
axe = figure.add_subplot()

sns.kdeplot(x=sepal_length, hue=iris.target, fill=True, ax=axe)
axe.set_xlabel(iris.feature_names[feature_chosen])
axe.legend(labels=iris.target_names.tolist())
```
