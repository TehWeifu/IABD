Siguiendo con el DataFrame anterior y usando Seaborn:

- Muestra el gráfico KDE de la distribución del largo del pétalo y del largo del sépalo.
- Muestra el gráfico KDE de la distribución del largo del pétalo y del largo del sépalo pero separado por el tipo de flor.
- Muestra un pairplot

```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['tipo_flor'] = iris.target
df
```

- Muestra un scatter plot del ancho del pétalo en el eje X y el ancho del sépalo en el eje Y según el tipo de flor.

```python
sns.scatterplot(df, x=df['petal width (cm)'], y=df['sepal width (cm)'], hue=df['tipo_flor'])
```

Muestra el gráfico KDE de la distribución del largo del pétalo y del largo del sépalo.

```python
figure = plt.figure(figsize=(16, 5))
axes = figure.add_subplot(1, 2, 1)
sns.kdeplot(x="petal length (cm)", data=df, fill=True, ax=axes)
axes = figure.add_subplot(1, 2, 2)
sns.kdeplot(x="sepal length (cm)", data=df, fill=True, ax=axes)
```

Muestra el gráfico KDE de la distribución del largo del pétalo y del largo del sépalo pero separado por el tipo de flor.

```python
figure = plt.figure(figsize=(16, 5))
axes = figure.add_subplot(1, 2, 1)
sns.kdeplot(x="petal length (cm)", data=df, fill=True, ax=axes, hue='tipo_flor')
axes = figure.add_subplot(1, 2, 2)
sns.kdeplot(x="sepal length (cm)", data=df, fill=True, ax=axes, hue='tipo_flor')
```

Muestra un pairplot

```python
pairplot = sns.pairplot(df, hue="tipo_flor")
```
