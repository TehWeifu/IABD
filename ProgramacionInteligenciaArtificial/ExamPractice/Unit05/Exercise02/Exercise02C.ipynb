Siguiendo con el DataFrame anterior:

- Haz una regresión lineal de los datos para cada tipo de activación.
- Haz un gráfico que:
    - Muestre la recta de la regresión de cada función de activación.
    - En la etiqueta de cada función de activación se muestre también el valor de R² para cada uno de ellos.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./tiempos_red_neuronal.csv', delimiter=',', index_col=0)
df
```

```python
df['tiempo'] = df['tiempo'] / 100
```

```python
figure = plt.figure(figsize=(14, 14))
axe = figure.add_subplot()
figure.suptitle('Regresion lineal del tiempo de entrenamiento de una red neuronal', color='blue', fontsize=18)
axe.set_xlabel('Epoca (Nº)', color='blue')
axe.set_ylabel('Tiempo (s)', color='blue')

sns.scatterplot(data=df, ax=axe, x='talla', y='tiempo', hue='funcion_activacion')
```
