Ejercicio 16: plot_surface

Dibuja una superficie en 3D , sabiendo que:
$$x \in [-15,15]$$

$$y \in [-15,15]$$
$$z =\frac {sin(\sqrt {x^2+y^2+4})}{\sqrt {x^2+y^2+4}}$$

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import LightSource
```

```python
x = np.linspace(-15, 15, 1_000)
y = np.linspace(-15, 15, 1_000)
x, y = np.meshgrid(x, y)

z = np.divide(
    np.sin(np.sqrt(x ** 2 + y ** 2 + 4)),
    np.sqrt(x ** 2 + y ** 2 + 4)
)
```

```python
light_source = LightSource()
facecolors = light_source.shade(z, plt.cm.cool, blend_mode='soft')

figure = plt.figure(figsize=(15, 12))
axe = figure.add_subplot(1, 1, 1, projection='3d')
axe.set_xlabel('X')
axe.set_ylabel('Y')
axe.set_zlabel('Z')


axe.plot_surface(x, y, z, facecolors=facecolors)

figure.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.cool), ax=axe, shrink=0.5)
```
