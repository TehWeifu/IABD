Usando la función plot_single_kde, modifica el gráfico anterior de forma que donde había los huecos, se llame a la función plot_single_kde

La gráfica debe quedar como la siguiente:

```python
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
```

```python
def plot_single_scatter(axes, x, y, target, label_x, label_y):
    scatter = axes.scatter(x, y, c=target, cmap='brg')

    axe.set_xlabel(label_x)
    axe.set_ylabel(label_y)
    axe.legend(handles=scatter.legend_elements()[0], labels=iris.target_names.tolist())
```

```python
def plot_single_kde(axes, x, target, label_x):
    sns.kdeplot(x=x, hue=target, fill=True, ax=axes)
    axe.set_xlabel(label_x)
    axe.legend(iris.target_names.tolist())
```

```python
iris = load_iris()
```

```python
grid_size = iris.data.shape[1]

figure = plt.figure(figsize=(20, 20))

for i in range(grid_size):
    for j in range(grid_size):
        axe = figure.add_subplot(grid_size, grid_size, (grid_size * i + j + 1))

        if i == j:
            plot_single_kde(axe, iris.data[:, i], iris.target, iris.feature_names[i])
        else:
            plot_single_scatter(axe, iris.data[:, i], iris.data[:, j], iris.target,
                                iris.feature_names[i], iris.feature_names[j])
```
