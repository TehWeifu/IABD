```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
iris = load_iris()
```

```python
def print_graph(x, y, axes):
    axes.set_facecolor("#F0F7FF")
    axes.grid(visible=True, which='major', axis='both', color="#FFFFFF", linewidth=1)
    axes.set_axisbelow(True)

    scatter = axes.scatter(iris.data[:, x], iris.data[:, y], c=iris.target)

    axes.set_xlabel(iris.feature_names[2], fontsize=15, labelpad=20, color="#003B80")
    axes.set_ylabel(iris.feature_names[0], fontsize=15, labelpad=20, color="#003B80")

    # Set x and y limits
    axes.set_xlim([0, 10])
    axes.set_ylim([0, 10])

    # Set x and y ticks
    # axes.set_xticks(np.arange(0, 11, 1))  #create ticks from minimum to maximum with step 1
    # axes.set_yticks(np.arange(0, 11, 1))  #create ticks from minimum to maximum with step 1
    axes.xaxis.set_major_locator(MaxNLocator(integer=True))
    axes.yaxis.set_major_locator(MaxNLocator(integer=True))

    # Here we create legend
    labels = np.unique(iris.target)
    legends = axes.legend(handles=scatter.legend_elements()[0], labels=iris.target_names.tolist())
```

```python
figure, arr_axes = plt.subplots(
    ncols=4,
    nrows=4,
    figsize=(16, 16),
    layout="constrained"
)

for i in range(iris.data.shape[1]):
    for j in range(iris.data.shape[1]):
        print_graph(i, j, arr_axes[i][j])

figure.suptitle("Flower features correlation", fontsize=50)
plt.show()
```
