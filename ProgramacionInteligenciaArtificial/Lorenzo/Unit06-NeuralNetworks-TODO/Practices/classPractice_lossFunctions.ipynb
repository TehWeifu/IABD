```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
losses = np.arange(0.01, 0.99, 0.01)
```

```python
def absolute_1_value(value):
    return abs(1 - value)
```

```python
def absolute_0_value(value):
    return abs(0 - value)
```

```python
def square_1_value(value):
    return (1 - value) ** 2
```

```python
def square_0_value(value):
    return (0 - value) ** 2
```

```python
def binaryCrossentropy_1_value(values):
    to_return = []
    for value in values:
        to_return.append(1 * log(value) + (1 - 1) * log(1 - value))
    return np.array(to_return) * -1
```

```python
def binaryCrossentropy_0_value(values):
    to_return = []
    for value in values:
        to_return.append(0 * log(value) + (1 - 0) * log(1 - value))
    return np.array(to_return) * -1
```

```python
losses_absolute_1 = absolute_1_value(losses)
losses_absolute_0 = absolute_0_value(losses)
losses_square_1 = square_1_value(losses)
losses_square_0 = square_0_value(losses)
losses_crossentropy_1 = binaryCrossentropy_1_value(losses)
losses_crossentropy_0 = binaryCrossentropy_0_value(losses)
```

```python
figure = plt.figure(figsize=(15, 10))

axes1 = figure.add_subplot(3, 2, 1)
axes1.plot(losses, losses_absolute_1)
axes1.set_title("|1 - ŷ|")
axes1.set_xlabel("ŷ")
axes1.set_ylabel("loss")

axes2 = figure.add_subplot(3, 2, 2)
axes2.plot(losses, losses_absolute_0)
axes2.set_title("|0 - ŷ|")
axes2.set_xlabel("ŷ")
axes2.set_ylabel("loss")

axes3 = figure.add_subplot(3, 2, 3)
axes3.plot(losses, losses_square_1)
axes3.set_title("(1 - ŷ)²")
axes3.set_xlabel("ŷ")
axes3.set_ylabel("loss")

axes4 = figure.add_subplot(3, 2, 4)
axes4.plot(losses, losses_square_0)
axes4.set_title("(0 - ŷ)²")
axes4.set_xlabel("ŷ")
axes4.set_ylabel("loss")

axes5 = figure.add_subplot(3, 2, 5)
axes5.plot(losses, losses_crossentropy_1)
axes5.set_title("binary crossentropy (y=1)")
axes5.set_xlabel("ŷ")
axes5.set_ylabel("loss")

axes6 = figure.add_subplot(3, 2, 6)
axes6.plot(losses, losses_crossentropy_0)
axes6.set_title(
    r"$- \frac{1}{N} \sum_{i=1}^{N} y_{i} \cdot log(\hat{y_i}) + (1-y_{i}) \cdot log(1-\hat{y_i}) $ when y=0")
axes6.set_xlabel("ŷ")
axes6.set_ylabel("loss")

figure.tight_layout()
```
