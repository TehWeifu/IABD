## Creando data frames
El objeto central de Pandas es el data frame. Es comparable a una "hoja de Excel" y es donde viven nuestros datos. Casi todo lo que vamos a hacer se orientará alrededor de crear un data frame y utilizar funciones del data frame para hacer nuestros análisis.

Lo fundamental es que un data frame tiene filas y columnas. Cada fila tiene un "nombre" que es el index y cada columna tiene un nombre que se refiere como la columna y ya está. Acedemos a los datos refiriéndonos a los índices y columnas.

### Como se crean?
Realmente hay varias formas de crear un data frame. Lo más común es tener un diccionario de listas donde los "keys" son las columnas y las listas son los datos de cada fila.

Pandas es bastante listo y acepta otros formatos. Lo fundamental es que tiene que entender que son las filas y las columnas. No vamos a cubrir todo porque realmente es más común cargar datos directamente (no crear los DFs a mano).

```python
data = {
    'columna_1': [1, 2, 3, 4, 5],
    'columna_2': [6, 7, 8, 9, 10],
    'columna_letras': ['a', 'b', 'c', 'd', 'e'],
}

my_df = pd.DataFrame(data)

my_df
```

```python
data = [
    (1, 6, 'a'),
    (2, 7, 'b')
]

pd.DataFrame(data, columns=['columna_1', 'columna_2', 'columna_letras'])
```

## Vectorización
Algo muy especial de Pandas es que aprovecha de la vectorización de operaciones de Numpy. En Python los loops son lentos y los debemos de evitar. Entonces, ¿cómo podemos hacer una operación a todas las filas de nuestro data set? Pues simplemente aprovechamos (siempre cuando sea posible) las funciones vectorizadas

Ya hemos visto algunos ejemplos

```python
# Dividir toda la columna por 1000 sin loop
house_data.SalePrice / 1000
```

```python
# Ratio entre 2 columnas
house_data.SalePrice / house_data.LotArea
```

¡Hay muchos ejemplos! Con operaciones básicas casi siempre hay una funciona vectorizada y merece la pena buscar. Podemos llegar a hacer cosas complejas...

```python
((
         (house_data.SalePrice / 1000) /
         (house_data.LotArea - house_data.loc[house_data.Neighborhood.str.contains('Ames').fillna(False), 'GarageArea'])
 ) ** 2).dropna()
```

## Creando visualizaciones
Data frames muy guay, pero no es siempre la mejor forma de transmitir conclusiones.

Una visualización se compone por unas piezas importantes:

* Un data frame
* Un(os) "aesthetics"
* Una(s) geometrías
* Un(os) temas o "themes"

Todos los gráficos se construyen a partir de estas piezas y nos permite crear visualizaciones simples o complejas.

```python
data = pd.DataFrame({
    'columna_1': [1, 2, 3, 4, 5],
    'columna_2': [6, 7, 8, 9, 10],
    'columna_letras': ['a', 'a', 'd', 'd', 'e'],
})

graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_point()
)

graph.draw()
```

```python
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_point() +
        pn.theme(axis_text=pn.element_text(size=20))
)

graph.draw()
```

## Las geometrías más comunes

```python
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_line()
)

graph.draw()
```

```python
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_col()
)

graph.draw()
```

```python
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1'})) +
        pn.geom_histogram(binwidth=4)
)

graph.draw()
```

## Otras cosas...

```python
# cambiar las etiquetas
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_point() +
        pn.xlab('Mejor etiqueta para la X')
)

graph.draw()
```

```python
# Combinar geometrías
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_line() +
        pn.geom_point()
)

graph.draw()
```

```python
# Darle un poco de estilo
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2'})) +
        pn.geom_point(color='red', size=10, shape='x')
)

graph.draw()
```

```python
# Color por variables
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2', 'color': 'columna_letras'})) +
        pn.geom_point(size=10)
)

graph.draw()
```

```python
# Color por variables
graph = (
        pn.ggplot(data, pn.aes(**{'x': 'columna_1', 'y': 'columna_2', 'color': 'columna_letras'})) +
        pn.geom_line(size=5) +
        pn.geom_point(size=10)
)

graph.draw()
```
