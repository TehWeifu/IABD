### Data cleaning

When we create a model, there is no way of "magically" tell the model what we want. We have to use data that represent properly our problem and create a model that predicts something which fits the answer that we are looking for.

If the data is "dirty" (bad, with issues, reflects something that does not represent our problem well), our model will predict things that don't make a lot o sense, they are not trustworthy or just plain wrong.

Things that we want to avoid:
* Data that have nothing to do with our problem
* Data that have a low sample size
* Outliers
* Nulls

1. Identify the main issues with the variables that seem (for now) the most interesting for the model.

The variables which seem the most relevant for the model have the following issues:
    - UsefulArea: has a pretty big standard deviation.
    - Neighborhood: uneven sample quantity between neighborhoods. Categorical data, must be transformed into numerical values.
    - OverallCond: data distribution is rather low, value 5, represent most of the houses. Might be correlated with variable 'OverallQual'.
    - OverallQual: some values have little to no representation. Might be correlated with variable 'OverallCond'.
    - YearRemodAdd: Has a pretty big outlier on its lowest value '1950'.
    
It is also important to note the target variable 'SalePrice' has pretty big outliers, and it would be convenient to handle them before introducing them to the model

2. Fix the issues

```python
# Encoding 'Neighborhood' to sorted numerical values given the mean of their sale prices
house_data_group_price = house_data.groupby("Neighborhood")["SalePrice"].mean().reset_index()
house_data_group_price_sorted = house_data_group_price.sort_values(by="SalePrice")
neighborhoods_sorted = house_data_group_price_sorted["Neighborhood"].tolist()
neighborhood = house_data[['Neighborhood']]
encoder = OrdinalEncoder(categories=[neighborhoods_sorted])
encoded_neighborhood = encoder.fit_transform(neighborhood)
house_data['Neighborhood'] = encoded_neighborhood
```

```python
# Clipping year remodeled (or built) to at last 1950
house_data['YearRemodAdd'] = house_data['YearRemodAdd'].clip(lower=1950)
```

```python
# Clipping the sale price to handle outliers 
lower_percentile = house_data['SalePrice'].quantile(0.05)
upper_percentile = house_data['SalePrice'].quantile(0.95)
house_data['SalePrice'] = house_data['SalePrice'].clip(lower_percentile, upper_percentile)
```

3. Delete all the data that are not relevant for the problem. Simplify the data.

Deleting rows from the dataset that do not represent regular a household sale (as explained above).

```python
house_data.drop(house_data[house_data["SaleCondition"] != "Normal"].index, inplace=True)
house_data.drop(house_data[house_data["SaleType"] != "WD"].index, inplace=True)
```

### Variable relevance

- Having fewer of the most relevant variables is usually better, because it simplifies the model.

To analyze the relevance, we usually look for "correlations" between different variables and the target. There are many ways of analyzing these "correlations", such as:
- Correlation coefficient (continuous variables)
- The mean of the target for each value of the variables (NON-continuous variables)

### Train and test
Splitting the data into train and test (or trains, test and validation) is a fundamental step for the correct evaluation of our model. We seek to replicate the effect of "obtaining new data".

The features chosen for the final `X` are: UsefulArea, Neighborhood, OverallQual, YearRemodAdd


```python
features_chosen = [
    "UsefulArea",
    "Neighborhood",
    "OverallQual",
    "YearRemodAdd",
]
target = "SalePrice"

X = house_data[features_chosen]
y = house_data[target]
```

2. Split the data into train / test

```python
YEAR_SOLD_VALIDATION_CUTOFF = 2_009
YEAR_SOLD_TEST_CUTOFF = 2_010

# Split data into train, validation and test
train_df = house_data[house_data['YrSold'] < YEAR_SOLD_VALIDATION_CUTOFF]
validation_df = house_data[
    (house_data['YrSold'] >= YEAR_SOLD_VALIDATION_CUTOFF) & (house_data['YrSold'] < YEAR_SOLD_TEST_CUTOFF)
    ]
test_df = house_data[house_data['YrSold'] >= YEAR_SOLD_TEST_CUTOFF]

X_train = train_df[features_chosen]
X_validation = validation_df[features_chosen]
X_test = test_df[features_chosen]

y_train = train_df[target]
y_validation = validation_df[target]
y_test = test_df[target]

# Print split results
print(f"Splitting the dataset:")
print(
    f"\t• Train dataset (<{YEAR_SOLD_VALIDATION_CUTOFF}) has a total of {len(train_df)} records. Represent {(len(train_df) / len(house_data) * 100):.2f}% of the total data."
)

print(
    f"\t• Validation dataset ({YEAR_SOLD_TEST_CUTOFF}) has a total of {len(validation_df)} records. Represent {(len(validation_df) / len(house_data) * 100):.2f}% of the total data."
)

print(
    f"\t• Test dataset (>={YEAR_SOLD_TEST_CUTOFF}) has a total of {len(test_df)} records. Represent {(len(test_df) / len(house_data) * 100):.2f}% of the total data."
)
```

### Evaluate the model
We are not going to spend a lot of effort in this step for now, but it is essential to evaluate the predictions we have generated. The goal is to understand how good our model is, to know if we can trust the results and use the to answer our problem.

1. Choose the metric for your evaluation

The metrics chosen to evaluate the model are: 
- Mean absolute error
- Median absolute error
- Root mean squared error

They were chosen because they are the most common metrics for linear regression problems overall.

2. Evaluate the predictions generated

```python
# Printing to console metrics about prediction results with validation dataset

print("Validation dataset metrics")
print()

mean_validation_sale_price = y_validation.mean()
print(f"Mean price is ${mean_validation_sale_price:,.2f}")

baseline = np.empty(len(y_validation))
baseline.fill(mean_validation_sale_price)
baseline_mean_absolute_error = metrics.mean_absolute_error(y_validation, baseline)
baseline_median_absolute_error = metrics.median_absolute_error(y_validation, baseline)
baseline_root_mean_squared_error = metrics.mean_squared_error(y_validation, baseline, squared=False)

print()
print("Model baseline")
print(
    f"Mean absolute error: ${baseline_mean_absolute_error:,.2f} ({(baseline_mean_absolute_error / mean_validation_sale_price * 100):,.2f}% off)")
print(
    f"Median absolute error: ${baseline_median_absolute_error:,.2f} ({(baseline_median_absolute_error / mean_validation_sale_price * 100):,.2f}% off)")
print(
    f"Root Mean squared error: ${baseline_root_mean_squared_error:,.2f} ({(baseline_root_mean_squared_error / mean_validation_sale_price * 100):,.2f}% off)")

metric_mean_absolute_error = metrics.mean_absolute_error(y_validation, y_predict)
metric_median_absolute_error = metrics.median_absolute_error(y_validation, y_predict)
metric_root_mean_squared_error = metrics.mean_squared_error(y_validation, y_predict, squared=False)

print()
print("Model metrics")
print(
    f"Mean absolute error: ${metric_mean_absolute_error:,.2f} ({(metric_mean_absolute_error / mean_validation_sale_price * 100):,.2f}% off)")
print(
    f"Median absolute error: ${metric_median_absolute_error:,.2f} ({(metric_median_absolute_error / mean_validation_sale_price * 100):,.2f}% off)")
print(
    f"Root mean squared error: ${metric_root_mean_squared_error:,.2f} ({(metric_root_mean_squared_error / mean_validation_sale_price * 100):,.2f}% off)")
```

```python
# Create a dispersion graph to compare predictions vs perfect predictions and baseline predictions

metrics_df = pd.DataFrame(data={
    "target": y_validation / 1_000,
    "prediction": y_predict / 1_000,
    "Area (m²)": X_validation['UsefulArea'] * 0.092903,
    "baseline": baseline / 1_000
})

graph = (
        pn.ggplot(
            metrics_df,
            pn.aes(**{"x": "prediction", "y": "target", "color": 'Area (m²)'}),
        ) +
        pn.geom_point() +
        pn.geom_line(pn.aes(**{"y": 'prediction'}), color='red') +  # Perfect prediction
        pn.geom_line(pn.aes(**{"y": 'baseline'}), color='blue') +  # Baseline prediction
        pn.labs(  # Actual predictions            
            title='Price predicted vs real price',
            x='Predicted price (x1000 $)',
            y='Real price (x1000 $)'
        ) +
        pn.theme(
            rect=pn.element_rect(color="black", size=1, fill='#aaa'),
            plot_title=pn.element_text(size=15, face="bold", hjust=.5),
        )
)

graph.draw()
```
