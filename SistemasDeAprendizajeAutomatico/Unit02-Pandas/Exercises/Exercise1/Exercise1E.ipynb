Siguiendo con el DataFrame anterior y usando Seaborn:


```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sample_data_flowers = load_iris()
columns = sample_data_flowers.feature_names
# print(dir(sample_data_flowers))
# print(sample_data_flowers.data)

data_frame_flowers: pd.DataFrame = pd.DataFrame(sample_data_flowers.data, columns=columns)
data_frame_flowers['flower type'] = sample_data_flowers.target
```

Muestra un scatter plot del ancho del pétalo en el eje X y el ancho del sépalo en el eje Y según el tipo de flor.

```python
figure = plt.figure(figsize=(12, 8))
axes = figure.add_subplot()

sns.scatterplot(x="petal width (cm)", y="sepal width (cm)", hue="flower type", data=data_frame_flowers, ax=axes)
```

Muestra el gráfico KDE de la distribución del largo del pétalo y del largo del sépalo.

```python
figure = plt.figure(figsize=(16, 5))
axes = figure.add_subplot(1, 2, 1)
sns.kdeplot(x="petal length (cm)", data=data_frame_flowers, fill=True, ax=axes)
axes = figure.add_subplot(1, 2, 2)
sns.kdeplot(x="sepal length (cm)", data=data_frame_flowers, fill=True, ax=axes)
```

Muestra el gráfico KDE de la distribución del largo del pétalo y del largo del sépalo pero separado por el tipo de flor.


```python
figure = plt.figure(figsize=(16, 5))
axes = figure.add_subplot(1, 2, 1)
sns.kdeplot(x="petal length (cm)", hue="flower type", data=data_frame_flowers, fill=True, ax=axes)
axes = figure.add_subplot(1, 2, 2)
sns.kdeplot(x="sepal length (cm)", hue="flower type", data=data_frame_flowers, fill=True, ax=axes)
```

Muestra un pairplot

```python
pairplot = sns.pairplot(data_frame_flowers, hue="flower type")
```
