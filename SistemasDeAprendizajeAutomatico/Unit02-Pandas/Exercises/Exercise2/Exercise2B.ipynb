```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

model_metrics = pd.read_csv("tiempos_red_neuronal.csv", sep=",", index_col=0)
model_metrics.rename(columns={'talla': 'epoca'}, inplace=True)
model_metrics = model_metrics.dropna()
model_metrics.drop(columns=["desviacion"], inplace=True)
model_metrics.head()
```

Muestra un scatter plot
 - Eje X:Época
 - Eje Y:Tiempo
 - Separa los datos por colores según la función de activación
 - Añade un título al gráfico y a los 2 ejes.

```python
figure = plt.figure(figsize=(12, 8))
axes = figure.add_subplot()

sns.scatterplot(x="epoca", y="tiempo", hue="funcion_activacion", data=model_metrics, ax=axes)
plt.title('Metricas red neuronal')
plt.xlabel('Número de épocas')
plt.ylabel('Tiempo entrenamiento (s)')
plt.show()
```
